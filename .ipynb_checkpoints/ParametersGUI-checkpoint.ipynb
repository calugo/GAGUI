```python
def datafr(A):
    
    gl=list(set(A['generation'].values))
    gns=['gen'+str(k).rjust( len(gl[0])-3 ,'0') for k in range(len(gl))]
    
    prs=[]
    B=list(A)
    B.remove('generation')
    B.remove('Fitness')
    Fmax=max(list(A["Fitness"]))
    pairs={}
    n=0
    for k in list(B):
        #print(k)
        pairs[n]=[]
        pairs[n].append(k)
        for j in B:
            if j!=k:
                pairs[n].append(j)
              
        B.remove(k)
        n+=1
        #print(B)
    cases=['All']
    for n in pairs.keys():
        zu=pairs[n][0]
        for m in pairs[n]:
            if zu!=m:
                cases.append(zu+'-'+m)
    #print(cases)
    #print(pairs)
    #Check for differences
    #z = set(A['generation']).difference(set(gns))
    #print(z)
    
    a = widgets.FloatSlider(value=Fmax,
    min=0.0,
    max=Fmax,
    step=0.1,
    description='Fitness:',
    disabled=False,
    continuous_update=False,  
    readout_format='.1f',
    )
    
    
    b = widgets.Button( description='Parameters-Plots', disabled=False,
        button_style='', # 'success', 'info', 'warning', 'danger' or ''
        tooltip='Click me')
    c = widgets.Dropdown(options=gns, value=gns[0], description='Generation:', disabled=False,)
    
    d = widgets.Dropdown(options=cases, value=cases[0], description='Parameters:', disabled=False,)

        
    def f(a,c,d):
        #print(c,d)
        if d=='All':
            #palette = sns.color_palette("viridis")#, length)
            #sns.set_palette(palette)
            AP=A.loc[(A['generation']==c) & (A['Fitness']<=a)]
            g=sns.pairplot(AP, hue="Fitness", diag_kws=dict(hue=None),corner=True, palette="viridis")#.set_title("LaLaLa")
            g.fig.suptitle(c, y=1.08)
        #plt.title('Training: %s' % gna)
            plt.show()
        else:
            vrx=d.split("-")
            print(vrx)
            APS=A.loc[(A['generation']==c) & (A['Fitness']<=a)]
            g=sns.relplot(x=vrx[0], y=vrx[1], hue="Fitness", data=APS,palette="viridis")
            g.fig.suptitle(c, y=1.08)
            plt.show()
    
    def dropdown_eventhandler(change):
        #with
        f(change.new)
        print(change.new)
    
    ui = widgets.HBox([a,c,d])
    out = widgets.interactive_output(f, {'a':a,'c': c, 'd': d})
    display(ui,out)
    #c.observe(dropdown_eventhandler, names='value')
```

```python
datafr(GR)
```
